```python
import pandas as pd
path="G:/resources/数据库/pro1/release of project 1/data/student_course.csv"
df=pd.read_csv(path)
```

```python
cols = ['studentId', 'course1', 'course2','course3','course4','course5','course6']
cols=['student_id','course_id']
cols1 = ['studentId', 'course1']
cols2 = ['studentId', 'course2']
cols3 = ['studentId', 'course3']
cols4 = ['studentId', 'course4']
cols5 = ['studentId', 'course5']
cols6 = ['studentId','course6']
df1=pd.DataFrame(columns=cols)
df2=pd.DataFrame(columns=cols)
df3=pd.DataFrame(columns=cols)
df4=pd.DataFrame(columns=cols)
df5=pd.DataFrame(columns=cols)
df6=pd.DataFrame(columns=cols)
df1[cols]=df[cols1]
df2[cols]=df[cols2]
df3[cols]=df[cols3]
df4[cols]=df[cols4]
df5[cols]=df[cols5]
df6[cols]=df[cols6]
df7=pd.DataFrame(columns=cols)
```

```python
df7=pd.concat([df1,df2,df3,df4,df5,df6],axis=0)
```

```python
df7=df7.dropna()
```

```python
df7=df7.drop_duplicates()
```

```python
df7
```

```python
df8=pd.DataFrame(columns=['student_id','course_id','class_id'])
df8['student_id']=df7['student_id']
df8['course_id']=df7['course_id']
```

```python
df8['course_id'].isnull().sum()
```

```python
path3="G:/resources/数据库/pro1/release of project 1/data/class.csv"
df_class=pd.read_csv(path3)
```

```python
df_class
```

```python
s_c1=pd.merge(df7,df_class,how='left',on=u'course_id')
```

```python
s_c1=s_c1.drop_duplicates(subset=['student_id','course_id'])
```

```python
s_c1
```

```python

s_c2=pd.DataFrame(columns=['student_id','class_id'])
s_c2['student_id']=s_c1['student_id']
s_c2['class_id']=s_c1['id']
#s_c2['class_id']=s_c2.class_id.apply(lambda x:int(x))
s_c2=s_c2.dropna(subset=['class_id'])
s_c2=s_c2.drop_duplicates(subset=['student_id','class_id'])
```

```python
s_c2['class_id']=s_c2.class_id.apply(lambda x:int(x))
```
