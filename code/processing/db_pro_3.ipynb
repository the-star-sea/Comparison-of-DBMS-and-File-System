```python
path1="course_info.json"
df_course=pd.read_json(path1)

col1=['courseId','courseName']
col2=['course_id','prerequisite_id1','prerequisite_id','kind']
df_c1=pd.DataFrame(columns=col1)
df_c2=pd.DataFrame(columns=col2)
df_c1[col1]=df_course[col1]
df_c1=df_c1.drop_duplicates()
df_c1[50:100]
df_c2['course_id']=df1['course_id']
df_c2['prerequisite_id1']=df1['prerequisite_id']
df_c2['kind']=df1['kind']
```

```python
def find_id(x):
    k=0
    for i in df3[df3['name']==x]['id'].values:
        if i!=0:
            k=i
    return k

df_c2['prerequisite_id']=df_c2.prerequisite_id1.apply(lambda x:find_id(x))
```

```python
df_c2['prerequisite_id']=df_c2.prerequisite_id.apply(lambda x:None if x==0 else x)
df_c2=df_c2.dropna()
```
